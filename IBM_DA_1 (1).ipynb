SUMMARY OF FEATURES:

---
The dataset contains employee personal details such as age, gender, marital status, and education.

It includes job details like department, job role, job level, and job involvement.

The data shows work conditions, such as overtime, business travel, and distance from home.

It provides salary and compensation information including monthly income, daily rate, hourly rate, salary hike, and stock options.

Employee experience details are included, such as years at the company, years in the current role, years with the same manager, and years since last promotion.

The dataset shows past work experience, including the number of companies previously worked.

It contains employee ratings like performance rating, work–life balance, job satisfaction, relationship satisfaction, and environment satisfaction.

The data includes employee identification and department-wise employee count.

It also shows working hours satisfaction and whether the company hires replacements for employees who leave.

Overall, the dataset gives a complete view of employee background, work life, experience, and satisfaction


EXPLORATORY DATA ANALYSIS:
---

EDA is a process of segmentation of features and segregating into groups.In this process we have segmented into Income based Analysis,Behavioural and Satisfaction analysis,Experience Based Analysis and Working Based Analysis.

```python
#Income based Analysis
# 1 What is the average salary of the employee for per day?
#datafrname.groupby("columnnme1")["columname2"].agg()
avg_salary=data.groupby("Department")["DailyRate"].mean()
avg_salary
```

```python
#VISUALIZATION FOR AVG SALARY OF EMPLOYEE
plt.bar(avg_salary.index,avg_salary.values)
plt.title("Average Salary of an Employee")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.show()
```

OBSERVATION:

---
From the above visualization of average salary of an employee:

1.There are three departments in the IBM company,which is HR,Research and development and sales.

2.The highest average salary is Research department and lowest average salary for a day is HR department

3.Research and sales department get almost same salary when comparing to the other.


```python
#2 What is highest salary for an hour?
#datafrname.groupby("columname1")["columname2"].agg()
highest_salary=data.groupby("Department")["HourlyRate"].max()
highest_salary
```
